```python
df = results.merge(rank[["country_full", "total_points", "previous_points", "rank_date"]], 
                   left_on=["date", "home_team"], 
                   right_on=["rank_date", "country_full"]).drop(["rank_date", "country_full"], axis=1)

df = df.merge(rank[["country_full", "total_points", "previous_points", "rank_date"]], 
              left_on=["date", "away_team"], 
              right_on=["rank_date", "country_full"], suffixes=("_home", "_away")).drop(["rank_date", "country_full"], axis=1)

df = df.drop(["tournament", "city", "neutral", "country"], axis=1)
```

```python
def swap_columns(df, col1, col2):
    tmp = df[col1].copy(deep=True)
    df[col1] = df[col2]
    df[col2] = tmp
    return df
```

```python
df_swaped = df.copy(deep=True)
df_swaped = swap_columns(df_swaped, "home_team", "away_team")
df_swaped = swap_columns(df_swaped, "home_score", "away_score")
df_swaped = swap_columns(df_swaped, "total_points_home", "total_points_away")
df_swaped = swap_columns(df_swaped, "previous_points_home", "previous_points_away")
```

```python
df = pd.concat([df, df_swaped], ignore_index=True).sort_values("date").reset_index(drop=True)
df
```

```python
def feature_maker(date, team1, team2):
    res = []
    team1_data = df[(df["date"] <= date) & (df["home_team"] == team1)]
    team2_data = df[(df["date"] <= date) & (df["home_team"] == team2)]

    if len(team1_data) < 6 or len(team2_data) < 6:
        return False

    res.append(team1_data.previous_points_home.iloc[-1]/1000)
    res.append(team2_data.previous_points_home.iloc[-1]/1000)

    for i in range(-2, -7, -1):
        res.append(team1_data.home_score.iloc[i])
        res.append(team1_data.away_score.iloc[i])
        res.append(team1_data.total_points_away.iloc[i]/1000)

    for i in range(-2, -7, -1):
        res.append(team2_data.home_score.iloc[i])
        res.append(team2_data.away_score.iloc[i])
        res.append(team2_data.total_points_away.iloc[i]/1000)

    return res
```

```python
from tqdm import tqdm
```

```python
train_x, train_y = [], []
for idx, date, team1, team2, score1, score2, total1, prev1, total2, prev2 in tqdm(df.itertuples(), total=len(df)):
    feature = feature_maker(date, team1, team2)
    if not feature:
        continue
    else:
        train_x.append(feature)
        if score1 == score2:
            train_y.append([0, 1, 0])
        elif score1 > score2:
            train_y.append([1, 0, 0])
        else:
            train_y.append([0, 0, 1])
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
from keras.layers import Dense, Dropout
from keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
train_x, train_y = np.array(train_x), np.array(train_y)

train_x, valid_x, train_y, valid_y = train_test_split(train_x, train_y, test_size = 0.2, shuffle=True, stratify=train_y, random_state=37)
```

```python
model = tf.keras.models.Sequential([
    Dense(128, input_shape=(32,), activation='relu'),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(16, activation='relu'),
    Dense(3, activation='softmax')
])

model.summary()

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['acc'])

checkpoint = ModelCheckpoint(filepath='checkpoint.ckpt',
                             save_weights_only=True,
                             save_best_only=True,
                             monitor='val_loss',
                             verbose=1)

early_stop = EarlyStopping(monitor='val_loss', patience=10)

model.fit(train_x, train_y,
          validation_data=(valid_x, valid_y),
          epochs=10000,
          callbacks=[checkpoint, early_stop],
          batch_size=16,
          shuffle=True,
          verbose=1)

model.load_weights('checkpoint.ckpt')
```

# World Cup Simulation

```python
dfs = pd.read_html(r"https://en.wikipedia.org/wiki/2022_FIFA_World_Cup#Teams")
```

```python
from collections.abc import Iterable

for i in range(len(dfs)):
    df_ = dfs[i]
    cols = list(df_.columns.values)
    
    if isinstance(cols[0], Iterable):
        if any("Tie-breaking criteria" in c for c in cols):
            start_pos = i+1

        if any("Match 46" in c for c in cols):
            end_pos = i+1
```

```python
matches = []
groups = ["A", "B", "C", "D", "E", "F", "G", "H"]
group_count = 0 

table = {}
#TABLE -> TEAM, POINTS, WIN PROBS (CRITERIO DE DESEMPATE)
table[groups[group_count]] = [[a.split(" ")[0], 0, []] for a in list(dfs[start_pos].iloc[:, 1].values)]

for i in range(start_pos+1, end_pos, 1):
    if len(dfs[i].columns) == 3:
        team_1 = dfs[i].columns.values[0]
        team_2 = dfs[i].columns.values[-1]
        
        matches.append((groups[group_count], team_1, team_2))
    else:
        group_count+=1
        table[groups[group_count]] = [[a, 0, []] for a in list(dfs[i].iloc[:, 1].values)]
```

```python
from operator import itemgetter
```

```python
advanced_group = []
last_group = ""

for k in table.keys():
    for t in table[k]:
        t[1] = 0
        t[2] = []
        
for teams in matches:
    draw = False

    feature = np.array([feature_maker(pd.to_datetime("2022-11-30"), teams[1], teams[2])])
    result = model.predict(feature, verbose=0)
    
    prob1, prob_draw, prob2 = result[0, 0], result[0, 1], result[0, 2]
    

    if max(prob1, prob_draw, prob2) == prob_draw:
        draw=True
        for i in table[teams[0]]:
            if i[0] == teams[1] or i[0] == teams[2]:
                i[1] += 1
                
    elif max(prob1, prob_draw, prob2) == prob1:
        winner = teams[1]
        winner_proba = prob1
        for i in table[teams[0]]:
            if i[0] == teams[1]:
                i[1] += 3
                
    elif max(prob1, prob_draw, prob2) == prob2:  
        winner = teams[2]
        winner_proba = prob2
        for i in table[teams[0]]:
            if i[0] == teams[2]:
                i[1] += 3
    
    for i in table[teams[0]]: #adding criterio de desempate (probs por jogo)
            if i[0] == teams[1]:
                i[2].append(prob1)
            if i[0] == teams[2]:
                i[2].append(prob2)

    if last_group != teams[0]:
        if last_group != "":
            print("\n")
            print("Group %s advanced: "%(last_group))
            
            for i in table[last_group]: #adding crieterio de desempate
                i[2] = np.mean(i[2])
            
            final_points = table[last_group]
            final_table = sorted(final_points, key=itemgetter(1, 2), reverse = True)
            advanced_group.append([final_table[0][0], final_table[1][0]])
            for i in final_table:
                print("%s -------- %d"%(i[0], i[1]))
        print("\n")
        print("-"*10+" Starting Analysis for Group %s "%(teams[0])+"-"*10)
        
        
    if draw == False:
        print("Group %s - %s vs. %s: Winner %s with %.2f probability"%(teams[0], teams[1], teams[2], winner, winner_proba))
    else:
        print("Group %s - %s vs. %s: Draw"%(teams[0], teams[1], teams[2]))
    last_group =  teams[0]

print("\n")
print("Group %s advanced: "%(last_group))

for i in table[last_group]: #adding crieterio de desempate
    i[2] = np.mean(i[2])
            
final_points = table[last_group]
final_table = sorted(final_points, key=itemgetter(1, 2), reverse = True)
advanced_group.append([final_table[0][0], final_table[1][0]])
for i in final_table:
    print("%s -------- %d"%(i[0], i[1]))
```

```python
advanced = advanced_group
playoffs = {"Round of 16": [], "Quarter-Final": [], "Semi-Final": [], "Final": []}
```

```python
for p in playoffs.keys():
    playoffs[p] = []

actual_round = ""
next_rounds = []

for p in playoffs.keys():
    if p == "Round of 16":
        control = []
        for a in range(0, len(advanced*2), 1):
            if a < len(advanced):
                if a % 2 == 0:
                    control.append((advanced*2)[a][0])
                else:
                    control.append((advanced*2)[a][1])
            else:
                if a % 2 == 0:
                    control.append((advanced*2)[a][1])
                else:
                    control.append((advanced*2)[a][0])

        playoffs[p] = [[control[c], control[c+1]] for c in range(0, len(control)-1, 1) if c%2 == 0]
        
        for i in range(0, len(playoffs[p]), 1):
            game = playoffs[p][i]
            
            home = game[0]
            away = game[1]

            feature = np.array([feature_maker(pd.to_datetime("2022-11-30"), home, away)])
            result = model.predict(feature, verbose=0)
            
            prob1, prob_draw, prob2 = result[0, 0], result[0, 1], result[0, 2]

            
            if actual_round != p:
                print("-"*10)
                print("Starting simulation of %s"%(p))
                print("-"*10)
                print("\n")
            
            if prob1 < prob2:
                print("%s vs. %s: %s advances with prob %.2f"%(home, away, away, prob2))
                next_rounds.append(away)
            else:
                print("%s vs. %s: %s advances with prob %.2f"%(home, away, home, prob1))
                next_rounds.append(home)
            
            game.append([prob1, prob2])
            playoffs[p][i] = game
            actual_round = p
        
    else:
        playoffs[p] = [[next_rounds[c], next_rounds[c+1]] for c in range(0, len(next_rounds)-1, 1) if c%2 == 0]
        next_rounds = []
        for i in range(0, len(playoffs[p])):
            game = playoffs[p][i]
            home = game[0]
            away = game[1]

            feature = np.array([feature_maker(pd.to_datetime("2022-11-30"), home, away)])
            result = model.predict(feature, verbose=0)
            
            prob1, prob_draw, prob2 = result[0, 0], result[0, 1], result[0, 2]   

            if actual_round != p:
                print("-"*10)
                print("Starting simulation of %s"%(p))
                print("-"*10)
                print("\n")
            
            if prob1 < prob2:
                print("%s vs. %s: %s advances with prob %.2f"%(home, away, away, prob2))
                next_rounds.append(away)
            else:
                print("%s vs. %s: %s advances with prob %.2f"%(home, away, home, prob1))
                next_rounds.append(home)
            game.append([prob1, prob2])
            playoffs[p][i] = game
            actual_round = p
```
